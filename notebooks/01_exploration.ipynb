```python
import pandas as pd

df = pd.read_csv('../data/Course_Completion_Prediction.csv')
```

```python
print("="*70)
print(" DATA TYPES SUMMARY")
print("="*70)

# Create a detailed info dataframe
data_info = pd.DataFrame({
    'Column': df.columns,
    'Data Type': df.dtypes.values,
    'Non-Null Count': df.notnull().sum().values,
    'Null Count': df.isnull().sum().values,
    'Null %': (df.isnull().sum().values / len(df) * 100).round(2),
    'Unique Values': df.nunique().values
})

display(data_info)

print("\n" + "="*70)
print(" DATA TYPES DISTRIBUTION")
print("="*70)
print(df.dtypes.value_counts())
```

```python


# categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
categorical_cols = ['Course_Name', 'Category']


for col in categorical_cols:
    print(f"\n{'='*50}")
    print(f" {col}")
    print('='*50)
    print(df[col].value_counts())
    print(f"Nombre total de données dans la colonne {col} : {df[col].value_counts().sum()}")
```

```python
CourseByCategorie = df.groupby(['Category', 'Course_Name']).size().reset_index(name='Count')

total = CourseByCategorie['Count'].sum()


print(CourseByCategorie)

print(f"Nombre total de données : {total}")
```
